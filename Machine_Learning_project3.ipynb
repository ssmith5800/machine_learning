```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import style
style.use("ggplot")
```

```python
df = pd.read_csv('Datasets/employee_attrition_train.csv')
```

```python
df.head()
```

```python
df1 = df.dropna()
```

```python
X = df1[['Age', 'DistanceFromHome', 'OverTime', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'StandardHours', 'TotalWorkingYears', 'WorkLifeBalance', 
         'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
X = pd.get_dummies(X)
y = df1['Attrition']

print("Shape: ", X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
pd = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
pd.head(20)
```
